````python
#importing the libararies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
````

````python
#load the data from the csv
def load_data():
    df = pd.read_csv('epa-sea-level.csv')
    return  df
````

````python
#draw scatter plot
def scatterplot():    
    df = load_data()
    ax = plt
    ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], color= 'red')

````

````python
#line reg for all data
def regression():
    df = load_data()
    ax = plt
    fit  = np.polyfit(df['Year'],df['CSIRO Adjusted Sea Level'], 1)
    fit_fn = np.poly1d(fit)
    ax.plot(df['Year'], fit_fn(df['Year']), color="blue")

````

````python
#regression from the year => 2000  

def regression2():
    df = load_data()
    ax = plt
    df_recent = df[df['Year'] >= 2000]
    fit_recent = np.polyfit(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'], 1)
    fit_fn_recent = np.poly1d(fit_recent)
    ax.plot(df_recent['Year'], fit_fn_recent(df_recent['Year']), color='green')
````

````python
#draw a plot for the regression
def draw_plot():
    
    fig = plt
    fig.subplots(figsize=(14, 7))

    #create a scatter plot of the data
    scatterplot()

    #create a line for best fit ,linear regression for all data 
    regression()
    
    # Create a second line of best fit starting from 2000
    regression2()

    #show the plot
    fig.savefig("regression.png")
    return fig
````
